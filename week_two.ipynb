#Function that grabs images from a webpage 


```
import mimetypes

def print_images(url, start=None, stop=None):
  results= requests.get(url)

  soup= BeautifulSoup(results.text , "html.parser")

  for img in soup.find_all("img")[start:stop]:
    if img["src"].startswith("//"):
      full_img = "https:" + img["src"]
    else:
      full_img= "https://en.wikipedia.org " +img["src"]
    mimetype=mimetypes.guess_type(full_img)
    if mimetype[0] is None:
      continue
    if mimetype[0].startswith("image"): 
      display(Image(full_img, width=100))

url= input("Paste a web address here : ")
print_images(url, 1, 5)   


```

#Copy

You can see where tese point in memory using the id()function 

here they point to the same area in memory 

```
data1=5
data2=data1

print(id(data1), id(data2))
```

Once we change one of values only that one will change, in python primitives are immutable and any time we change the value it really is just giving it a new value so we are going to have a change in the memory address 

```
data1=5
data2=data1

print(id(data1), id(data2))

data1=6

print(data1, data2)
print(id(data1), id(data2))
```

When working with objects we are basically creating an allias , data2 points to data1. They are still going to have the same address in memory.  
Primitives are immutuable- if you change them they are actually replaced.  
Objects are different because you can change the data in object and the memory address stayes the same.   

```
data1=[5]
data2=data1

print(id(data1), id(data2))

data1[0]=6

print(data1, data2)
print(id(data1), id(data2))
```

#Deep Copy


```
from copy import deepcopy

def print_list(data):
  for d in data:
    print(id(d))
  print()  

data1=['0', 1 ,  2, 3, [4], {5}, {6:7}]
data2=deepcopy(data1)

print_list(data1)
print_list(data2)

data1[4][0]= 'new'
print(data1, data2)

print_list(data1)
print_list(data2)
```

Here it is changing the value in the object but not the object itself.   
You can only change the data in objects , you cannot replace them with new objects when you working wtih functions.

#pprint

pprint is used to make nice outputs for dictionaries

```
from pprint import pprint

grades = {
    
    1: {
        'math': [90, 90, 30, 20],
        'science': [94, 93, 92, 91],
        'history': [12, 23, 34, 54],
        'reading': [54, 23, 23, 54]
    },

    2: {
        'math': [9, 9, 3, 2],
        'science': [9, 9, 9, 9],
        'history': [1, 2, 3, 5],
        'reading': [5, 2, 2, 5]
        
    },

    3: {
        'math': [45, 30, 20, 0],
        'science': [4, 3, 2, 41],
        'history': [42, 33, 34, 54],
        'reading': [54, 23, 33, 54]
        
    }
}
pprint(grades)
```

```
import json

print(json.dumps(grades, indent=2))
```

Write a json file 

```
#pass in a data file i.e data.json, it is just a text file with a json extension, then use string w to say we are writing in this file 
#and give that file a name i.e f 
with open('data.json', 'w') as f:
  #use dump to write to a file , it will be in our files 
  json.dump(grades, f) 
```

writing it nicer way:

```
with open('data.json', 'w', encoding= 'utf-8') as f:
  json.dump(grades, f, ensure_ascii= False, indent=4) 
```

Dictionary Read from the file.  


You can use json file to store configurations  , then you could change the json file in your code and when you are done you could save it back to the json file and then next time you lunch the application read the json file to keep the configuration you had from the last execution 

```
with open('data.json') as f:
  grades= json.load(f)

pprint(grades)  
```
